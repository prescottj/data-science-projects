### Challenge 1.1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers.

```python
# Create a list of each combination of digits1 and digits2
import itertools
digits1 = range(100,1000)
digits2 = range(100,1000)
d1_d2=list(itertools.product(digits1,digits2))

# Create a dictionary where the key is d1_d2 and the value contains the product, the reverse of the product, and whether it is a palindrome
product_dict = {}
for row in d1_d2:
    product=str(row[0]*row[1])
    rev_product=product[::-1]
    if product==rev_product:
        palindrome=True
    else:
        palindrome=False
    product_dict[row[0:2]]=product,rev_product,palindrome

# Subset dictionary to values which are palindromes
subset_dict=dict((k,v) for k,v in product_dict.items() if v[2]==True)
value1=[]
for k,v in subset_dict.items():
    value1.append(int(v[0]))

# Output dictionary entries for largest palindrome
largest_palindrome=dict((k,v) for k,v in subset_dict.items() if v[0]==str(max(value1)))
largest_palindrome
```


### Challenge 1.2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000.

```python
import numpy as np

# Find unique list of composite numbers within the range 2 - 1999
composite=[]
for n in range(2,2000):
    for d in range(2,n):
        if n % d ==0:
            composite.append(n)
composite = np.unique(composite)
        
#print "Set of unique composite numbers within range:", composite

# The list of prime numbers will be the remaining numbers in the range 2 - 1999
prime=[]
for n in range(2,2000):
    if n not in composite:
        prime.append(n)
        
#print "List of prime numbers within range:", prime

# Sum all numbers within the generated list of prime numbers
sum_prime = sum(prime)
print "Sum of prime numbers:", sum_prime
```

### Challenge 1.3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000.

```python
# Create a list of all natural numbers below 1000 that are divisible by 3 or 5
divisible3_5=[]
for n in range(1,1000):
    if n%3==0 or n%5==0:
        divisible3_5.append(n)
        
# Sum over the list of numbers
sum_divisible=sum(divisible3_5)
print "Sum of natural numbers below 1000 that are multiples of 3 or 5:", sum_divisible
```

### Challenge 3: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other.

```python
# THIS SOLUTION IS CASE SENSITIVE

string_of_interest=raw_input("Which string would you like to compress? ")

import re
def compression(string_of_interest):
    # Split sequence each time a character changes
    split_char=[m.group(0) for m in re.finditer(r"(\D)\1*", string_of_interest)]
    
    # Create empty lists for the count of each iteration of characters and which character is being counted
    char_count=[]
    unique_char=[]
    
    # Iterating over each sequence to populate the count and character lists
    for s in split_char:
        char_count.append(len(s))
        unique_char.append(list(set(s)))
    
    # Convert the counts from integer to string elements separated by commas
    str_char_count=str(char_count).split(', ')
    
    # List comprehension to combine list of character lists into a single list
    lst_unique_char=[l[0] for l in unique_char]
    
    # Merge each character with corresponding count into a tuple
    comb_list=zip(lst_unique_char, str_char_count)
    
    # Merge elements within tuple for entire list
    merge_list=[]
    for elements in comb_list:
        merge_list.append(''.join(elements))
    
    # Merge each element within list into a single string
    single_string=''.join(merge_list)
    
    # Remove brackets from the final string
    final_string=single_string.replace('[','').replace(']','')
    
    # Return the original string if string cannot be compressed, the compressed string otherwise
    can_be_compressed = next((True for item in char_count if item > 1), False)
    if can_be_compressed==True:
        final_string=final_string
    else:
        final_string=string_of_interest
    
    print 'Compressed string:', final_string
    
compression(string_of_interest)
```
